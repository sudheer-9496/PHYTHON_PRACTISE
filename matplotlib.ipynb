```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([30,25,25,20])

y=["CSE", "Mech", "Civil" , "ECE"]

plt.pie(x, labels =y)

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([9,10,11,12])

y=np.array(["maths", "english", "python" , "beee"])

plt.plot(x, y)

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([0.0,1.0,2.0,3.0])

y=np.array([3,8,1,10])

plt.plot(x,y,marker="o",ls=":")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([30,25,25,20])

y=["Python", "chemistry", "English" , "Maths"]

z=[0,0,0,0.2]

plt.pie(x, labels = y, explode = z)

plt.show()
```
